## Here is an explanation of all the data categories found in bootstrap-static (General Data):

# elements
Master list of all the datsets. 88 total

# element_stats
Focuses on player stats from the 'elements' master list

# element_types
Player position-specific. Contains codes, descriptions, and other data related to broad positional items (max on field at once, mi of field at once, etc)

# events
specific fantasy league data relating to that week's games, wrt running the league (rg-deadlines for submissions, if it's the current week, best score, etc.)

# game_settings
Sets the rules & configurations that govern how the league operates in season

# phases
Info about the season's phases (eg each month, first/last Gameweeks, etc.)

# teams
Detailed info about EPL team performance during the season (not the fantasy teams). Lots of statistical categories

# total_players
Absolute sum of the total players in the fantasy.premierleague league:  10,064,327


Endpoint	                                Description
/api/bootstrap-static/	                    General data (players, teams, gameweeks)
/api/element-summary/{element_id}/	        Individual player data (stats, fixtures)
/api/fixtures/	                            All Premier League fixtures
/api/event/{event_id}/live/	                Live gameweek data
/api/events/	                            Gameweek information (deadlines, averages)
/api/teams/	                                Information about Premier League teams
/api/dream-team/{gameweek}/	                Dream team for a specific gameweek
/api/leagues-classic/{league_id}/standings/	Classic league standings
/api/leagues-h2h/{league_id}/standings/	    H2H league standings
/api/entry/{team_id}/	                    Manager’s team details
/api/entry/{team_id}/event/{event_id}/picks/ Manager’s picks for a specific gameweek
/api/entry/{team_id}/transfers/	            Manager’s transfer history
/api/stats/elements/	                    Player ownership and stats
