```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
```

```python
fpl_defenders = pd.read_csv('../_defense_analysis/data/processed/integration_defense.csv')
sorare_defenders = pd.read_csv('../sorare/sorare_data/predictions/sorare_defenders_predictions.csv')

print(fpl_defenders.columns)
print(sorare_defenders.columns)

fpl_defenders = fpl_defenders.rename(columns={'first_name': 'First_Name', 'second_name': 'Last_Name', 'epl_predictions':'Predicted_Points'})
fpl_defenders_filtered = fpl_defenders[['id','First_Name','Last_Name', 'Predicted_Points']]

sorare_defenders_filtered = sorare_defenders[['First_Name','Last_Name', 'So_5_Scores_9', 'sorare_predictions']]

print(fpl_defenders.shape)
print(sorare_defenders.shape)

print("------------------------")

print(fpl_defenders_filtered.shape)
print(sorare_defenders_filtered.shape)
```

```python
defenders = pd.merge(sorare_defenders_filtered, fpl_defenders_filtered, on=['First_Name', 'Last_Name'] )

print(defenders.shape)

defenders['Weighted_BOR3_Score'] = (defenders['sorare_predictions'] * 0.75) + (defenders['Predicted_Points'] * 0.25)
```

```python
defenders = defenders.sort_values('Weighted_BOR3_Score', ascending=False)

defenders.to_csv('../data/predictions/weighted_bor3_defenders.csv')
defenders
```

```python
y_true = defenders['So_5_Scores_9'].to_numpy() 
y_pred = defenders['Weighted_BOR3_Score'].to_numpy()

# Calculate r2
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f'MSE: {mean_squared_error(y_true, y_pred)}')
print(f'RMSE: {rmse}')
r2 = r2_score(y_true, y_pred)
print(f'r2: {r2}')
```
